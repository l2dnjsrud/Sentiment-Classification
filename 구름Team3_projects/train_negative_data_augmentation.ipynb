```python
import nlpaug.augmenter.word as naw
```

```python
aug = naw.ContextualWordEmbsAug(model_path='bert-base-uncased', action="insert")
augmented_text = aug.augment(random_train_neg)
```

```python
augmented_text[:10]
```

```python
import pickle
```

```python
# pkl 형태로 저장
with open('augmented_train_neg.pkl', 'wb') as f:
    pickle.dump(augmented_text, f)
```

```python
# 저장된 pkl 불러오기
with open('augmented_train_neg.pkl','rb') as f:
    mydict = pickle.load(f)
```
